```python
#. How can you create a new column 'IsMinor' and set it to True for passengers under 18 using .loc[]?
# Convert 'Age' to numeric, forcing errors to NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Set 'IsMinor' using .loc[]
df.loc[df['Age'] < 18, 'IsMinor'] = True
df.loc[df['Age'] >= 18, 'IsMinor'] = False
```
